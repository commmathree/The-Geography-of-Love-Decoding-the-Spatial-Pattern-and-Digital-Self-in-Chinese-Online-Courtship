## Formal Scrapping Session:

```python
import lxml.html as html
import csv
from time import sleep
from random import randint
from time import time
from IPython.core.display import clear_output
#create all the column, 20 fields in total
ids=[]
genders=[]
ages=[]
stas=[]
locs1=[]
locs2=[]
heights=[]
weights=[]
edus=[]
salarys=[]
carB=[]
houseB=[]
agesR=[]
heightsR=[]
edusR=[]
stasR=[]
locsR=[]
picsR=[]
hukous=[]
intros=[]
```

```python
start_time = time()
requests = 0
```

```python
pages = [str(i) for i in range(178005000,178007000)]
```

```python
for page in pages:
        try:
          # Make a get request
            response = driver.get('http://www.jiayuan.com/' + page)
            # Pause the loop
            sleep(randint(8,15))

            # Parse the content of the request with BeautifulSoup
            html = driver.page_source
            soup = BeautifulSoup(html,"lxml")

            userID=soup.find("h4")
            ID=userID.span.text
            ID=re.findall(r'\d+',ID)[0]
            ids.append(ID)

            #gender
            gender=soup.find("ul",class_="nav_l").find('li')
            if gender.text.find("她")== -1:
                genderCode=1
            else:
                genderCode=0
            genders.append(genderCode)

            info1=soup.find("h6", class_="member_name")
            number=re.findall(r'\d+',info1.get_text())
            age=int(number[0])
            ages.append(age)

            selfintro=soup.find("div", class_="js_text").text
            intros.append(selfintro)
            
            status=info1.get_text()
            middle= status.split('，')[1]
            stas.append(middle)

            city=soup.find("h6", class_="member_name").find_all('a')
            loc1=city[0].text
            loc2=city[1].text
            locs1.append(loc1)
            locs2.append(loc2)

            height=soup.find("ul", class_="member_info_list fn-clear").find_all('li')[1].find("em")
            heights.append(height.text)
            
            weight=soup.find("ul", class_="member_info_list fn-clear").find_all('li')[5].find("em")
            weights.append(weight.text)
            
            edu=soup.find("ul", class_="member_info_list fn-clear").find_all('li')[0].find("em")
            edus.append(edu.text)
            
            salary=soup.find("ul", class_="member_info_list fn-clear").find_all('li')[3].find("em")
            salarys.append(salary.text)

            car=soup.find("ul", class_="member_info_list fn-clear").find_all('li')[2].find("em")
            if car.text=='已经购车':
                carCode=1
#             elif car.text=='--':
#                 carCode=2
            else:
                carCode=0
            carB.append(carCode)
            
            house=soup.find("ul", class_="member_info_list fn-clear").find_all('li')[4].find("em")
            if house.text=='已购住房':
                houseCode=1
#             elif house.text=='--':
#                 houseCode=2
            else:
                houseCode=0
            houseB.append(houseCode)
 
            ageR=soup.find("ul", class_="js_list fn-clear").find_all('li')[0].find("div","ifno_r_con")
            agesR.append(ageR.text)
           
            heightR=soup.find("ul", class_="js_list fn-clear").find_all('li')[1].find("div","ifno_r_con")
            heightsR.append(heightR.text)
           
            eduR=soup.find("ul", class_="js_list fn-clear").find_all('li')[3].find("div","ifno_r_con")
            edusR.append(eduR.text)
           
            statusR=soup.find("ul", class_="js_list fn-clear").find_all('li')[5].find("div","ifno_r_con")
            stasR.append(statusR.text)
            
            locR=soup.find("ul", class_="js_list fn-clear").find_all('li')[6].find("div","ifno_r_con_1")
            locsR.append(locR.text)

            picR=soup.find("ul", class_="js_list fn-clear").find_all('li')[4].find("div","ifno_r_con")
            picsR.append(picR.text)

            hukou=soup.find_all("ul", class_="js_list fn-clear")[5].find_all('li')[1].find('em')
            hukous.append(hukou.text)
            
            print('pageid： ', page,'Finished!')
            
        except: 
            pass 
```
